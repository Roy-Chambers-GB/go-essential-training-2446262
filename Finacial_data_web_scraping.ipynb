```python
# @title iShares iBoxx $ High Yield Corporate Bond ETF (HYG)
import requests
from bs4 import BeautifulSoup

url = "https://www.google.com/finance/quote/HYG:NYSEARCA"
response = requests.get(url)
response.raise_for_status()

soup = BeautifulSoup(response.content, "html.parser")

# Find the element containing the heading
heading_element = soup.find("div", class_="zzDege")

# Extract the heading text
heading_text = heading_element.text if heading_element else None

# Find the element containing the quote
quote_element = soup.find("div", class_="YMlKec fxKbKc")

# Extract the text content and convert to float if possible
quote_text = quote_element.text if quote_element else None
# Remove dollar sign and commas before converting to float
quote_value = float(quote_text.replace("$", "").replace(",", "")) if quote_text else None

print(f"{heading_text}: {quote_value}")
```

```python
# @title Shanghai Composite (SSEC)

# Defining Trading Limits
import datetime

def get_limits():
  """Gets the low and high limits from a file or prompts for input if the file is outdated."""
  today = datetime.date.today()
  try:
    with open("SSEC_trading_limits.txt", "r") as f:
      date_str, low_str, high_str = f.readline().strip().split(",")
      date = datetime.datetime.strptime(date_str, "%Y-%m-%d").date()
      if date == today:
        return float(low_str), float(high_str)
  except FileNotFoundError:
    pass  # File doesn't exist, proceed to get user input

  # Get user input for new limits
  low_limit = float(input("Enter the low trading limit: "))
  upper_limit = float(input("Enter the upper trading limit: "))

  # Store the new limits with the current date
  with open("SSEC_trading_limits.txt", "w") as f:
    f.write(f"{today.strftime('%Y-%m-%d')},{low_limit},{upper_limit}\n")

  return low_limit, upper_limit

# Get the current trading limits
low_limit, upper_limit = get_limits()

# Now you can use these limits in your trading logic
print(f"Low Limit: {low_limit}")
print(f"Upper Limit: {upper_limit}")


import matplotlib.pyplot as plt

# Fetching the XAG/USD Price
import requests
from bs4 import BeautifulSoup
from lxml import etree

url = "https://uk.investing.com/indices/shanghai-composite"
response = requests.get(url)
response.raise_for_status()

soup = BeautifulSoup(response.content, "html.parser")

# Use lxml to parse and find with XPath
dom = etree.HTML(str(soup))
# The original XPath might be outdated or incorrect.
# This new XPath targets the span element containing the price based on its aria-label attribute.
price_title_element = dom.xpath('/html/head/title')

# Check if any elements were found
if price_title_element:
    price_title_element = price_title_element[0]  # Get the first element if found
    price_title_text = price_title_element.text
    print(price_title_text)
else:
    print("Price title element not found.")

# Use lxml to parse and find with XPath
dom = etree.HTML(str(soup))
price_element = dom.xpath('//*[@id="__next"]/div[2]/div[2]/div[2]/div[1]/div[1]/div[3]/div[1]/div[1]/div[1]')[0]

price_text = price_element.text
price_value = float(price_text.replace(",", "")) if price_text else None

print(price_value)
print(f"{price_title_text}: {price_value}")

# Assign price_value to quote_value before percentage calculation
quote_value = price_value  # Corrected line: Assign the extracted price

# Calculate the percentage of position within limits
if quote_value < low_limit:
    position_percentage = 0  # Below low limit
elif quote_value > upper_limit:
    position_percentage = 100  # Above upper limit
else:
    position_percentage = (quote_value - low_limit) / (upper_limit - low_limit) * 100

# Print the position percentage
print(f"Position Percentage within Limits: {position_percentage:.2f}%")

# Create the horizontal bar chart
fig, ax = plt.subplots(figsize=(8, 1))  # Adjust figure size as needed
ax.barh([0], [position_percentage], color='blue')  # Create a single bar

# Set chart limits and labels
ax.set_xlim(0, 100)  # Percentage range
ax.set_ylim(-0.5, 0.5)  # Single bar, adjust for visibility
ax.set_yticks([])  # Hide y-axis ticks
ax.set_xlabel('Position Percentage (%)')

# Add a vertical line to indicate the current position
ax.axvline(x=position_percentage, color='red', linestyle='--')

# Display the chart
plt.show()
```

```python
# @title Copper Continuous contract (HGW00  COMEX)
from logging import exception

# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as widgets
#from ipywidgets import Layout, VBox, HBox, Spacer
from IPython.display import display, clear_output
import datetime
import requests
from bs4 import BeautifulSoup

# Create output widgets
out_widgets = widgets.Output()
out_chart = widgets.Output()  # For the line chart
out_bar_chart = widgets.Output() # For the bar chart

# Set margin for out_chart and out_bar_chart
out_chart.layout.margin = "10px 0px 0px 0px"  # Bottom margin of 10px
out_bar_chart.layout.margin = "20px 0px 0px 0px"  # Bottom margin of 10px

# Create widgets for limits input (inside out_widgets)
with out_widgets:
    #low_limit_widget = widgets.FloatText(description="Low Limit:", value=4.87)
    #upper_limit_widget = widgets.FloatText(description="Upper Limit:", value=5.34)
    low_limit_widget = widgets.FloatText(description="Low Limit:", value=0.00)
    upper_limit_widget = widgets.FloatText(description="Upper Limit:", value=0.00)
    refresh_checkbox = widgets.Checkbox(
        value=False, description='Refresh Limits', disabled=False, indent=False
    )
    bullish_checkbox = widgets.Checkbox(
        value=False, description='Bullish', disabled=False, indent=False
    )
    bearish_checkbox = widgets.Checkbox(
        value=False, description='Bearish', disabled=False, indent=False
    )
    display(
        low_limit_widget,
        upper_limit_widget,
        refresh_checkbox,
        bullish_checkbox,
        bearish_checkbox,
    )
# Imports and Widget Setup end

# Integrate Google Drive storage
from google.colab import drive
drive.mount('/content/drive')

# Data file paths
historical_data_path = '/content/drive/My Drive/copper_historical_data.pkl'
trading_limits_path = '/content/drive/My Drive/Copper_trading_limits.txt'

# Initialize DataFrame (if it doesn't exist)
try:
    df_historical = pd.read_pickle(historical_data_path)
except FileNotFoundError:
    df_historical = pd.DataFrame(columns=['Timestamp', 'Heading', 'Quote', 'Percentage', 'Low Limit', 'Upper Limit'])


# Function to fetch copper price
def fetch_copper_price():
    url = "https://www.google.com/finance/quote/HGW00:COMEX"
    response = requests.get(url)
    response.raise_for_status()

    soup = BeautifulSoup(response.content, "html.parser")

    # Find the element containing the heading
    heading_element = soup.find("div", class_="zzDege")

    # Extract the heading text
    heading_text = heading_element.text if heading_element else None

    # Find the element containing the quote
    quote_element = soup.find("div", class_="YMlKec fxKbKc")

    # Extract the text content and convert to float if possible
    quote_text = quote_element.text if quote_element else None
    # Remove dollar sign and commas before converting to float
    quote_value = float(quote_text.replace("$", "").replace(",", "")) if quote_text else None

    return heading_text, quote_value # Return both values

# Function to get trading limits
def get_limits(refresh=False):
    """Gets the low and high limits, prioritizing widget values when refreshing."""
    today = datetime.date.today()
    file_path = trading_limits_path  # Path on Google Drive

    # Get sentiment from checkboxes
    if bullish_checkbox.value:
        sentiment = "Bullish"
    elif bearish_checkbox.value:
        sentiment = "Bearish"
    else:
        sentiment = "Neutral"


    if refresh:  # Prioritize refresh from widgets
        low_limit = low_limit_widget.value
        upper_limit = upper_limit_widget.value

        # Store the new limits with the current date
        # with open("Copper_trading_limits.txt", "w") as f:
        with open(file_path, "w") as f:
            f.write(f"{today.strftime('%Y-%m-%d')},{low_limit},{upper_limit},{sentiment}\n")
            f.flush()  # Explicitly flush the buffer

        return low_limit, upper_limit, sentiment  # Return refreshed values immediately

    else:  # Try to read limits from the file, update widgets if found and current
        try:
          with open(file_path, "r") as f:  # Use file_path for reading
                date_str, low_str, high_str, sentiment_str = f.readline().strip().split(",") # Read sentiment
                date = datetime.datetime.strptime(date_str, "%Y-%m-%d").date()
                if date == today:
                    # Update widget values from file
                    low_limit_widget.value = float(low_str)
                    upper_limit_widget.value = float(high_str)
                    # Update sentiment from file (checkbox logic should handle consistency)
                    if sentiment_str == "Bullish":
                        bullish_checkbox.value = True
                        bearish_checkbox.value = False
                    elif sentiment_str == "Bearish":
                        bullish_checkbox.value = False
                        bearish_checkbox.value = True
                    else:
                        bullish_checkbox.value = False
                        bearish_checkbox.value = False
                    return float(low_str), float(high_str), sentiment_str # Return sentiment here too
        except FileNotFoundError:
            pass  # If file not found, continue to use widget values below

    # If refresh not requested or file is missing/outdated, use widget values
    low_limit = low_limit_widget.value
    upper_limit = upper_limit_widget.value

    # Store these values in the file, potentially creating it if missing
    with open(file_path, "w") as f:
        f.write(f"{today.strftime('%Y-%m-%d')},{low_limit},{upper_limit},{sentiment}\n")

    return low_limit, upper_limit, sentiment  # Return widget values

# Function to create/update the plot
def create_plot(df, low_limit, upper_limit): # receives low_limit and upper_limit
    fig, ax1 = plt.subplots(figsize=(12, 6))
    ax1.plot(df_historical['Timestamp'], df_historical['Quote'], color='blue')
    ax1.set_xlabel('Timestamp')
    ax1.set_ylabel('Quote', color='blue')
    ax1.tick_params('y', labelcolor='blue')
    ax1.axhline(y=low_limit, color='green', linestyle='--', label='Low Limit') # visualise Low Limit line
    ax1.axhline(y=upper_limit, color='red', linestyle='--', label='Upper Limit') # visualise Upper Limit line
    ax1.legend()  # Show the legend to identify the limit lines
    ax2 = ax1.twinx()
    ax2.plot(df_historical['Timestamp'], df_historical['Percentage'], color='red')
    ax2.set_ylabel('Percentage', color='red')
    ax2.tick_params('y', labelcolor='red')
    plt.title('Copper Continuous Contract Price and Percentage Over Time')
    plt.grid(True)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()


# Function to update the chart and potentially refresh limits
def update_chart():
    global low_limit, upper_limit, sentiment, df_historical # Set global access

    low_limit, upper_limit, sentiment = get_limits(refresh=refresh_checkbox.value)

    if refresh_checkbox.value:
        print(f"Refreshed Limits: Low - {low_limit}, Upper - {upper_limit}")

    heading_text, quote_value = fetch_copper_price()
    print(f"{heading_text}: {quote_value}")

    if quote_value < low_limit:
        position_percentage = 0  # Below low limit
    elif quote_value > upper_limit:
        position_percentage = 100  # Above upper limit
    else:
        position_percentage = (quote_value - low_limit) / (upper_limit - low_limit) * 100

    print(f"Position Percentage within Limits: {position_percentage:.2f}%")

    # Append new data to DataFrame
    new_row = pd.DataFrame({'Timestamp': [datetime.datetime.now()],
                            'Heading': [heading_text],
                            'Quote': [quote_value],
                            'Percentage': [position_percentage],
                            'Low Limit': [low_limit],
                            'Upper Limit': [upper_limit]})
    new_row['Timestamp'] = pd.to_datetime(new_row['Timestamp'])  # Convert timestamp
    df_historical = pd.concat([df_historical, new_row], ignore_index=True)
    df_historical.to_pickle(historical_data_path)

    # Line Chart Update (within out_chart context)
    with out_chart:
        clear_output(wait=True)  # Clear before plotting
        create_plot(df_historical, low_limit, upper_limit)


    # Ensure only one sentiment checkbox is selected (using link)
    def on_checkbox_change(change):
        if change['name'] == 'value':
            if change['new']:  # Checkbox is checked
                # Uncheck the other checkbox
                other_checkbox = (
                    bearish_checkbox
                    if change['owner'] == bullish_checkbox
                    else bullish_checkbox
                )
                other_checkbox.value = False
            update_chart()  # Update chart after checkbox change

    bullish_checkbox.observe(on_checkbox_change)
    bearish_checkbox.observe(on_checkbox_change)

    # Determine bar color based on checkboxes
    if bullish_checkbox.value:
        bar_color = 'green'
        sentiment = "Bullish"
    elif bearish_checkbox.value:
        bar_color = 'red'
        sentiment = "Bearish"
    else:
        bar_color = 'grey'
        sentiment = "Neutral"

    print(sentiment)  # Print the sentiment


    # Bar Chart Update (within out_bar_chart context)
    with out_bar_chart:
        clear_output(wait=True)
        fig, ax = plt.subplots(figsize=(8, 1))
        ax.barh([0], [position_percentage], color=bar_color)  # Use bar_color
        ax.set_xlim(0, 100)
        ax.set_ylim(-0.5, 0.5)
        ax.set_yticks([])
        ax.set_xlabel('Position Percentage (%)')
        ax.axvline(x=position_percentage, color='red', linestyle='--')
        plt.show()

    # Store Copper price to Google Drive
    heading_text, quote_value = fetch_copper_price() # Get both heading and value
    file_path = '/content/drive/My Drive/copper_price.txt'
    with open(file_path, 'w') as f:
        f.write(f"{heading_text}: {quote_value}\n") # Store with heading
        f.write(f"Position Percentage within Limits: {position_percentage:.2f}%") # Store position percentage

    print(f"Copper price and position percentage stored to {file_path}")

# Path to your pickle file on Google Drive
#file_path = '/content/drive/My Drive/copper_historical_data.pkl'

# Load the data into a DataFrame
#df_historical = pd.read_pickle(file_path)

# Convert 'Timestamp' column to datetime objects if needed
#df_historical['Timestamp'] = pd.to_datetime(df_historical['Timestamp'])

# Create the time series plot with two y-axes
#fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot "Quote" on the left y-axis
#ax1.plot(df_historical['Timestamp'], df_historical['Quote'], color='blue')
#ax1.set_xlabel('Timestamp')
#ax1.set_ylabel('Quote', color='blue')
#ax1.tick_params('y', labelcolor='blue')

# Create a second y-axis on the right
#ax2 = ax1.twinx()

# Plot "Percentage" on the right y-axis
#ax2.plot(df_historical['Timestamp'], df_historical['Percentage'], color='red')
#ax2.set_ylabel('Percentage', color='red')
#ax2.tick_params('y', labelcolor='red')

# Set title and grid
#plt.title('Copper Continuous Contract Price and Percentage Over Time')
#plt.grid(True)
#plt.xticks(rotation=45, ha='right')
#plt.tight_layout()
#plt.show()

# Preview Pandas pickle file data
# Print the first 5 rows
#print(df_historical.head())

# Print the 'Quote' column
# print(df_historical['Quote'])
# print(df_historical['Low Limit'])

# Print data for a specific date
# print(df_historical[df_historical['Timestamp'] == '2025-04-24'])  # Replace with your desired date

# Export Pandas pickle file data to CSV file.
df_historical.to_csv('/content/drive/My Drive/copper_historical_data.csv', index=False)

# Code: display, update_chart call, observe.
# Display the output widgets
# display(out_widgets, out_chart)

# Display widgets and output areas
ui = widgets.VBox([out_widgets, out_chart, out_bar_chart])  # Arrange vertically
display(ui)

# Call update_chart initially and potentially on a schedule or event
update_chart()

# Observe the checkbox for changes
refresh_checkbox.observe(lambda change: update_chart(), names='value')
#bullish_checkbox.observe(lambda change: update_chart(), names='value')
#bearish_checkbox.observe(lambda change: update_chart(), names='value')
```
